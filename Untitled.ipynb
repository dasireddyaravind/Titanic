Handling missing values is essential for improving the quality of my data analysis. When dealing with missing values, I have opted for predictive imputation as my chosen method for this dataset.

In predictive imputation, I divided the data into two subsets: one with complete data and the other with missing values. Then, I built a linear regression model using the subset of the data that had no missing values. This model helped me predict and fill in the missing values in the other subset.

Using the linear regression model, I could estimate the missing values based on the relationships observed in the complete data. This approach allowed me to retain as much valuable information as possible, resulting in a more comprehensive and accurate dataset for my analysis.

By employing predictive imputation, I ensured that my analysis was based on a more complete dataset, enhancing the reliability of my results and allowing me to draw more meaningful insights from the data.


```python
from sklearn.linear_model import LinearRegression

# Subsetting with missing values
data_with_missing = data[data.isnull().any(axis=1)]

# Subsetting without missing values
data_without_missing = data[~data.isnull().any(axis=1)]

# Training dataset
X_train = data_without_missing[['SibSp']]
y_train = data_without_missing['Age']

# Creating and training the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predicting the missing values in the other subset
X_test = data_with_missing[['SibSp']]
predicted_age = model.predict(X_test)

# Assigning the predicted values to the missing values in the original dataset
data.loc[data_with_missing.index, 'Age'] = predicted_age
```

```python
data = data.drop(['Name', 'Cabin','Ticket','Cabin_Class'], axis=1)

```

Before generating the model, we conducted an exploration of the target variable, which is the 'Survived' variable, to understand its distribution and determine whether it is balanced or imbalanced.

```python
data['Survived'].value_counts()

counts = data['Survived'].value_counts()
# Create pie chart
labels = ['Not Survived', 'Survived']
plot.pie(counts, labels=labels, autopct='%1.1f%%')

# Add title
plot.title('Survival Distribution')

# Display the chart
plot.show()


```

After analyzing the 'Survived' variable, I observed that it is imbalanced, with approximately 62% of the records belonging to the 'Not Survived' class and only 38% to the 'Survived' class. This imbalance can potentially lead to biased predictions, where my model may be more inclined to predict the majority class (i.e., 'Not Survived') due to its higher prevalence in the dataset.

To address this issue and prevent my model from being biased towards the majority class, I decided to use the under-sampling method. Under-sampling involves randomly removing samples from the majority class ('Not Survived' in this case) to create a more balanced dataset with an equal representation of both classes. By balancing the dataset through under-sampling, I can ensure that my model learns from both classes more effectively and produces fair and accurate predictions for both survival outcomes. This approach allows me to build a more reliable and robust predictive model that avoids favoring one class over the other during predictions.

```python
X = data.drop('Survived', axis=1)  # Features
y = data['Survived']  # Target variable

# importing the necessary libraries
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X, y)

# Checking the shape of the under-sampled features
print(X_resampled.shape) 

# Checking the value counts of the under-sampled target variable
print(y_resampled.value_counts()) 

counts = y_resampled.value_counts()

# Creating pie chart
labels = ['Not Survived', 'Survived']
plot.pie(counts, labels=labels, autopct='%1.1f%%')

plot.title('Survival Distribution after Random Undersampling')

plot.show()



```

```python
new_data = pd.DataFrame(X_resampled, columns=X.columns)
new_data['Survived'] = y_resampled

from sklearn.ensemble import RandomForestClassifier

# Performing one-hot encoding on categorical columns
X_new_encoded = pd.get_dummies(new_data)

# Separate the features (X) and target variable (y) from the new encoded dataset
X_new = X_new_encoded.drop('Survived', axis=1)
y_new = X_new_encoded['Survived']

# Create a Random Forest classifier
rf_model = RandomForestClassifier(random_state=42)

# Fit the model on the entire dataset
rf_model.fit(X_new, y_new)
```


After balancing the target variable by using the under-sampling method, I proceeded to create a random forest model using the training data. The random forest model is a powerful ensemble learning technique that combines multiple decision trees to make predictions.

With the balanced training data, the random forest model was trained to learn the relationships and patterns between the features and the target variable ('Survived'). By utilizing a combination of decision trees, the model can capture complex interactions and provide more accurate predictions compared to individual decision trees.

Once the random forest model was trained on the balanced training data, I will use it to predict the values for the target variable in the test data. The test data contains samples with features but lacks the corresponding 'Survived' values. By using the trained model, I can make predictions for the 'Survived' outcome on the test data based on the patterns learned during training.

This predictive model will allow me to assess the survival probabilities for passengers in the test data based on the features provided. By evaluating the model's performance and accuracy on the test data, I can gain insights into its effectiveness and make informed decisions for future applications or analyses.


```python
test_data = test_data.drop(['Cabin', 'Name', 'Ticket'], axis=1)
test_data.head()

from sklearn.impute import SimpleImputer

# Creating an imputer object to fill missing values with the mean
imputer = SimpleImputer(strategy='mean')

# Filling missing values in the 'Age' column
test_data['Age'] = imputer.fit_transform(test_data[['Age']])

# Filling missing values in the 'Fare' column
test_data['Fare'] = imputer.fit_transform(test_data[['Fare']])

test_data.isnull().sum()
```

Before using the model to predict the target variable on the test data, I performed data preprocessing on the test data. The data preprocessing steps included handling missing values and removing unnecessary variables, similar to what was done during the creation of the model on the training data.

To handle missing values in the test data, I used the imputation method by filling the missing values with the mean of the corresponding feature. Imputing the mean ensures that the missing values are replaced with a representative value, enabling the model to make predictions on complete data.

Furthermore, I removed the same unnecessary variables from the test data that were previously excluded during the creation of the model on the training data. This includes variables like 'Cabin', 'Name', and 'Ticket', which do not significantly contribute to predicting the target variable ('Survived') and were deemed irrelevant for the analysis.

```python
# Preprocessing the test_data
test_data_encoded = pd.get_dummies(test_data)

# Making predictions using the trained model
predictions = rf_model.predict(test_data_encoded)

# Adding the 'Survived' column to the test_data DataFrame
test_data['Survived'] = predictions

# Filtering the 'PassengerId' and 'Survived' columns
results = test_data.loc[:, ['PassengerId', 'Survived']]
```

After using the random forest model to predict the target variable ('Survived') on the test data, the predicted results were compared with the actual results provided in the test data. This comparison allowed us to evaluate the accuracy of the predictive model.


```python

predictions = results['Survived']
actual = gender_submission['Survived']

# Calculating accuracy
accuracy = (predictions == actual).mean()

# Multiply by 100 to get the percentage
accuracy_percentage = accuracy * 100

print("Accuracy:", accuracy_percentage)
```


In this case, the model achieved an accuracy of 76%. This means that approximately 76% of the predictions made by the model matched the actual survival status of the passengers in the test data.
